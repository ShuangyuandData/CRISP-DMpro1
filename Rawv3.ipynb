## Read Data

```python
## read data
def readdata(path):
    """
    Read the data and show the basic info
    
    Input: path of the data file
    
    """
    tempcp1=pd.read_csv(path)
    tempcp2=tempcp1.rename(columns={'Unnamed: 0' : 'Index'})
    tempcp3=tempcp2.set_index('Index')
    
    print(tempcp3.shape)
    
    return tempcp3
```

## Correlation of Matrices

```python
def identify(df):
    """
    Identify the ratio of material1
    """
    C1t=C1.to_numpy()
    C1t2=(C1t-np.mean(C1t))/np.std(C1t)
    
    C2t=C2.to_numpy()
    C2t2=(C2t-np.mean(C2t))/np.std(C2t)
    
    frac=[]   # percentage of material 1
    numcorr=[]  # num

    for i in range(101):
        ex=i/100
        frac.append(ex)
    
        temp=C1t2*ex+C2t2*(1-ex)
        #print(ex, (1-ex))
        temp2=temp.reshape(-1)
    
        tempM=df.to_numpy()
        tempM2=(tempM-np.mean(tempM))/np.std(tempM)
        tempM3=tempM2.reshape(-1)
    
        num=np.corrcoef(temp2,tempM3)[1,0]
        numcorr.append(num)
        
        
    zippedlist=list(zip(frac, numcorr))
                        
    final=pd.DataFrame(zippedlist, columns=['frac','numcorr'])
    
    final2=final.sort_values(by=['numcorr'], ascending=False)
    
    perc=final2.frac.iloc[0]
                        
    return final2, perc
    
```

```python
## Mixed 1
final1, perc1=identify(M1)
print(perc1)
final1.head()
```

```python
## Mixed 2
final2, perc2=identify(M2)
print(perc2)
final2.head()
```

```python
## Mixed 3
final3, perc3=identify(M3)
print(perc3)
final3.head()
```

```python
## Mixed 2R
final2R, perc2R=identify(M2R)
print(perc2R)
final2R.head()
```

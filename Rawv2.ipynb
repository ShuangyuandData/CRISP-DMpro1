## Transform the raw data to blob data

```python
def transtoblob(df, thres):
    """
    transform the raw data to blob data
    
    Input:  df is the raw data
            thres is the threshold for the peak value
    """
    
    nrow, ncol=df.shape
    
    RI=[]
    RII=[]
    PV=[]

    ### all the points inside

    for i in range(1,nrow-1):  # row id
        for j in range(1,ncol-1): # col id
        
        
            dft=df.iloc[(i-1):(i+2),(j-1):(j+2)]
            dftm=dft.to_numpy()
            maxvalue=dftm.max()
        
            if dft.iloc[1,1] == maxvalue and maxvalue > thres:  
                PV.append(maxvalue)
                RI.append(dft.columns[1])
                RII.append(dft.index[1])
                
    ### top

    for j in range(1,ncol-2): # col id
        
        dft=df.iloc[0:2,(j-1):(j+2)]
        dftm=dft.to_numpy()
        maxvalue=dftm.max()
        
        if dft.iloc[0,1] == maxvalue and maxvalue > thres:  
            PV.append(maxvalue)
            RI.append(dft.columns[1])
            RII.append(dft.index[0])
            
            
    ### bottom
    
    for j in range(1,ncol-2): # col id
        
        dft=df.iloc[nrow-2:nrow,(j-1):(j+2)]
        dftm=dft.to_numpy()
        maxvalue=dftm.max()
        
        if dft.iloc[1,1] == maxvalue and maxvalue > thres:
            PV.append(maxvalue)
            RI.append(dft.columns[1])
            RII.append(dft.index[1])
            
            
    ### left
    
    for i in range(1,nrow-2): # row id
        
        dft=df.iloc[(i-1):(i+2),0:2]
        dftm=dft.to_numpy()
        maxvalue=dftm.max()
        
        if dft.iloc[1,0] == maxvalue and maxvalue > thres:  
            PV.append(maxvalue)
            RI.append(dft.columns[0])
            RII.append(dft.index[1])
            
            
    ### right
    for i in range(1,nrow-2): # row id
        
        dft=df.iloc[(i-1):(i+2),ncol-2:ncol]
        dftm=dft.to_numpy()
        maxvalue=dftm.max()
        
        if dft.iloc[1,1] == maxvalue and maxvalue > thres: 
            PV.append(maxvalue)
            RI.append(dft.columns[1])
            RII.append(dft.index[1])
            
            
    ### four points

    # topleft
    dff1=df.iloc[0:2,0:2]
    maxvalue=dff1.to_numpy().max()
    if dff1.iloc[0,0] == maxvalue and maxvalue >thres:
            PV.append(maxvalue)
            RI.append(dff1.columns[0])
            RII.append(dff1.index[0])
        
        
    # topright        
    dff2=df.iloc[0:2,ncol-2:ncol]
    maxvalue=dff2.to_numpy().max()
    if dff2.iloc[0,1] == maxvalue and maxvalue >thres:
            PV.append(maxvalue)
            RI.append(dff2.columns[1])
            RII.append(dff2.index[0])
        
        
        
    # bottomleft
    dff3=df.iloc[nrow-2:nrow, 0:2]
    maxvalue=dff3.to_numpy().max()
    if dff3.iloc[1,0] == maxvalue and maxvalue >thres:
            PV.append(maxvalue)
            RI.append(dff3.columns[0])
            RII.append(dff3.index[1])
        
    #bottomright
    dff4=df.iloc[nrow-2:nrow, ncol-2:ncol]
    maxvalue=dff4.to_numpy().max()
    if dff4.iloc[1,1] == maxvalue and maxvalue >thres:
            PV.append(maxvalue)
            RI.append(dff4.columns[1])
            RII.append(dff4.index[1])
            
    
    
    final=pd.DataFrame(RI, columns=['RI'])
    final['RII']=RII
    final['PV']=PV
    
    return final
```
